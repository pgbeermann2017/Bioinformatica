```python
def compara_secuencia(s1,s2):
    n=len(s1.seq)
    m=len(s2.seq)
    mat=np.zeros((m,n),dtype=int)
    print(mat.shape)
    for i in range(0,n):
        for j in range(0,m):
          if s1.seq[i:i+3]==s2.seq[j:j+3]:
            mat[j,i]=1
    x=np.arange(n)
    y=np.arange(m)
    XX,YY=np.meshgrid(x,y)
    plt.contour(XX,YY,mat)
    plt.show()
    return mat
```

```python
compara_secuencia(s1,s2)
```
