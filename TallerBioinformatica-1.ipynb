## Listas

Las listas son conjuntos mutables de elementos que pueden ser del mismo tipo o de tipos diferentes. Para declarar una lista se encierra entre corchetes cuadrados los elemetos separados por comas.



```python
bases=["A","C","G","T"]  #Declaración de una lista de elementos 
print(bases)             #Utilizar print para imprimir la variable
```

```python
bases                    #imprime la variable bases 
```

```python
bases[3]                #utilizar un índice para acceder a un elemento de la lista
```

```python
len(bases)             #len(cadena) calcula el número de elemento
```

```python
bases.append("U")      #La función append() agrega un elemento a la cadena
bases
```

```python
bases.remove("U")     #la función remove() elimina un elemento de la cadena
bases
```

```python
for x in bases:     # imprime cada elemento x de la lista bases
    print(x)
```

```python
nuc=bases.pop()      #elimina el último elemento de la lista y lo asigna a la variable nuc
```

```python
bases.append("T")
bases
```

### Cálculo de las posiciones en donde se encuentra la secuencia "ATG" y el número de veces en que esta se repite

```python
sec="ATG"
lista=[]
num=0
for i in range(len(cadena)):
    #print(cadena[i:i+3])
    if sec==cadena[i:i+3]:
        #print(i)
        lista.append(i)
        num+=1
print(lista,num)
```
