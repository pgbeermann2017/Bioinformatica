```python
##Random Walk
fig, ax = plt.subplots(figsize=(8, 8))
dx=np.random.randn(1000)
dy=np.random.randn(1000)
dx[0]=0
dy[0]=0
x=dx.cumsum()
y=dy.cumsum()
ax.plot(x,y,'-',markersize=5)
ax.plot([0,0],[-40,40],alpha=0.5)
ax.plot([-40,40],[0,0],alpha=0.5)
ax.grid(True)
plt.show()
```
